```python
mean = np.average(midparent_height)

fig, ax = plot_histogram(data = midparent_height, 
    bins = 30,
    x_label = "Midparent Height (Inches)",
    y_label= "Density",
    title = "Distribution of Midparent Heights", 
    density = True)
ax.axvline(x = mean, ymin=0, ymax=1, color = 'red')
ax.annotate(f"Mean: {np.round(mean,2)}", xy = (73,0.25))

fig.savefig("images/lecture3parenthist_with_mean.png")
```

```python
sd = np.std(midparent_height)

fig, ax = plot_histogram(data = midparent_height, 
    bins = 30,
    x_label = "Midparent Height (Inches)",
    y_label= "Density",
    title = "Distribution of Midparent Heights", 
    density = True)

ax.hlines(y = 0.3, xmin = mean, xmax = mean+sd, color = 'Green')
ax.vlines(x = mean, ymin = 0.295, ymax = 0.305, color = 'Green')
ax.vlines(x = mean+sd, ymin = 0.295, ymax = 0.305, color = 'Green')


ax.annotate(f"SD: {np.round(sd,2)}", xy = (mean + 0.3*sd ,0.305))

fig.savefig("images/lecture3parenthist_with_SD1.png")
```

```python
sd = np.std(midparent_height)

fig, ax = plot_histogram(data = midparent_height, 
    bins = 30,
    x_label = "Midparent Height (Inches)",
    y_label= "Density",
    title = "Distribution of Midparent Heights", 
    density = True)

ax.hlines(y = 0.3, xmin = mean, xmax = mean+sd, color = 'Green')
ax.vlines(x = mean, ymin = 0.295, ymax = 0.305, color = 'Green')
ax.vlines(x = mean+sd, ymin = 0.295, ymax = 0.305, color = 'Green')

ax.hlines(y = 0.20, xmin = mean, xmax = mean+3*sd, color = 'red')
ax.vlines(x = mean, ymin = 0.195, ymax = 0.205, color = 'red')
ax.vlines(x = mean+3*sd, ymin = 0.195, ymax = 0.205, color = 'red')


ax.annotate(f"SD: {np.round(sd,2)}", xy = (mean + 0.3*sd ,0.305))
ax.annotate("3 SD from Mean", xy = (mean + sd ,0.205))

fig.savefig("images/lecture3parenthist_with_SD2.png")
```
